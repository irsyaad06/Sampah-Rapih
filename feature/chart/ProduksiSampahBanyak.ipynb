```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named data_jenis_sampah with columns 'Jenis Sampah', 'Produksi Sampah', 'Persentase'
%store -r data_jenis_sampah
# Filter rows where 'Produksi Sampah' is greater than 5000
filtered_data = data_jenis_sampah[data_jenis_sampah['Produksi Sampah'].astype(float) > 5000]

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(filtered_data['Jenis Sampah'], filtered_data['Produksi Sampah'].astype(float))

# Add labels and title
plt.xlabel('Jenis Sampah')
plt.ylabel('Produksi Sampah (m)')
plt.title('Jenis Sampah dengan Produksi Lebih dari 5000 ')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.legend(title='Tahun 2020')

# Show the plot
plt.show()
```
